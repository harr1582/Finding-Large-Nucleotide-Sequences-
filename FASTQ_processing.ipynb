## Python code to iterate over FASTQ files

This function returns all names of sequences with more than 30 nucleotides. The function also returns the percentage of sequences that are more than 30 nucleotides long.


```python
from Bio import SeqIO
fastq_seq = SeqIO.parse(r'C:\Users\harr1\Desktop\sample_files\fastq\read2\Sample_R2.fastq', "fastq") 

def greater_than_30(file):
    long_l = []
    total_l = []
    for each in file:
        total_l.append(each)
        if len(each) > 30:
            long_l.append(each)
    for each in long_l:
        print(each.name)

    return ('% of sequences with more than 30 nucleotides:', (len(long_l)/len(total_l))*100)    
greater_than_30(fastq_seq)
```
